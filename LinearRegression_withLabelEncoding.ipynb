```
import requests
import pandas as pd
url = 'https://raw.githubusercontent.com/codebasics/py/master/ML/5_one_hot_encoding/Exercise/carprices.csv'
res = requests.get(url, allow_redirects=True)
with open('sales_team.csv','wb') as file:
    file.write(res.content)
temp = pd.read_csv('sales_team.csv')
```

```
temp
```

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder
```

```
dfle = temp
```

```
dfle = dfle.rename(columns={'Car Model': 'car_model'})
dfle
```

```
dummies = pd.get_dummies(dfle.car_model)
dummies
```

```
dfle
```

```
merged = pd.concat([dfle, dummies],axis=1)
```

```
merged
```

```
final = merged.drop(['car_model'],axis=1)
```

```
final
```

```
final = final.drop(['Mercedez Benz C class'], axis = 1)
```

```
final
```

```
X = final.drop(['Sell Price($)'], axis = 1)
```

```
y = final[['Sell Price($)']]
```

```
X
```

```
y
```

```
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```
model.fit(X,y)
```

```
model.predict(X)
```

```
model.predict([[86000, 7, 0, 1]])
```

```
model.predict([[45000,4,0,0]])
```

```
model.score(X,y)
```
